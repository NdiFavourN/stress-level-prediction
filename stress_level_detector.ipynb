```python
import pandas as pd 
import numpy as np
import joblib
```

```python

from sklearn.model_selection import train_test_split    #to split the dataset
from sklearn.preprocessing import StandardScaler        # to put all the various data on the same scale
from sklearn.linear_model import LinearRegression       # the linear regression model for us to be able to use
from sklearn.metrics import mean_squared_error, r2_score  #for error handling

print("Libraries imported successfully!")
```

```python
data = pd.read_csv('StressLevelDataset.csv')
#data=data.isnull().sum()
print('csv file loaded and cleaned successfully')
print(data.isnull().sum())
```

```python
x = data.iloc[:,:-1]
y = data.iloc[:,-1]
data.describe()
```

```python
# --- Feature Scaling ---
# It's a best practice for linear models.
# We'll fit the scaler on the training data only to avoid data leakage.
scaler = StandardScaler()
x_train ,x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Fit the scaler on the training data and transform it
x_train_scaled = scaler.fit_transform(x_train)
# Transform the test data using the *same* scaler fitted on the training data
x_test_scaled = scaler.transform(x_test)
```

```python
# Initialize the Linear Regression model
model = LinearRegression()
```

```python
# Train the model
model.fit(x_train_scaled, y_train)

print("Model training complete.")
```

```python
#Make predictions on the test set
y_pred = model.predict(x_test_scaled)

# Evaluate the model's performance
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Model Performance on Test Set:")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (R2): {r2:.2f}")
```

```python
# Save the trained model to a file
joblib.dump(model, 'model.joblib')

# Save the scaler to a file
joblib.dump(scaler, 'scaler.joblib')

print("Model and scaler have been saved as 'model.joblib' and 'scaler.joblib'")
```
